#### Bài 1: 
Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó. 

```python
#Cách 1:
a, b = map(int, input('Nhập chiều dài và chiều rộng hình chữ nhật: ').split())
if a <= 0 or b <= 0:
    dien_tich = 0
    chu_vi = 0
else:
    dien_tich = a * b
    chu_vi = 2 * (a + b)    
print(f"Diện tích: {dien_tich}")
print(f"Chu vi: {chu_vi}")
```

```python
#Cách 2:
def tinh_hcn(a, b):
    if a <= 0 or b <= 0:
        return 0, 0
    return a * b, 2 * (a + b)

a, b = map(int, input('Nhập chiều dài và chiều rộng: ').split())
dt, cv = tinh_hcn(a, b)
print(f"Diện tích: {dt}")
print(f"Chu vi: {cv}")
```

#### Bài 5: 
Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(int, input("Nhập số nguyên a, b, c: ").split())

delta = b**2 -4*a*c
if a == 0 and b == 0 and c == 0:
    print("Phương trình có vô số nghiệm")
else:
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm kép x = {x}")
    else:
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        print(f"Phương trình có hai nghiệm phân biệt: x1 = {x1}, x2 = {x2}")
```

```python
# Cach 2:
import cmath
def giai_bac_2(a, b, c):
    if a == 0: 
       
        return giai_bac_1(b, c)
    delta = b**2 - 4*a*c
    x1 = (-b + cmath.sqrt(delta)) / (2*a)
    x2 = (-b - cmath.sqrt(delta)) / (2*a)
    res1 = x1.real if x1.imag == 0 else x1 #res la result   .Ý nghĩa: Nếu phần ảo bằng 0, biến res sẽ chỉ lấy phần thực (ví dụ: biến (2+0j) thành số 2.0). Nếu có phần ảo (nghiệm phức), nó sẽ giữ nguyên cả cụm số phức
    res2 = x2.real if x2.imag == 0 else x2  #x1.real: Lấy phần thực của số phức.
                                            #x1.imag == 0: Kiểm tra xem phần ảo có bằng 0 hay không (tức là nghiệm đó thực chất là một số thực)
    return res1, res2
   
try:
    
    a, b, c = map(float, input('Nhap cac so a, b, c: ').split())
    print(f"Ket qua: {giai_bac_2(a, b, c)}")
except ValueError:
    print("Loi: Vui long nhap dung 3 so!")
```

#### Bài 8:
Giải hệ phương trình tuyến tính:
- ax + by = m
- cx + dy = n

```python
# Cach 1:
a, b, c, d, m, n = map(float, input('Nhập giá trị: ').split())
if a == c and a > 0 and c > 0:
    y = (m-n) / (b-d)
    x = (m - b*y) / a
    print(f'Phương trình có nghiệm duy nhất: x= {x} , y ={y}')
elif b == d and b > 0 and d > 0:
    x = (m-n) / (a-c)
    y = (m - a*x) / b
    print(f'Phương trình có nghiệm duy nhất: x= {x} , y ={y}')
elif a == b == c == d == 0 and m > 0 and n > 0:
    print('Phương trình có vô nghiệm')
else:
    print('Phương trình có vô số nghiệm')
```

```python
# Cach 2:

def he_phuong_trinh_chi_tiet(a, b, m, c, d, n):
    D  = a*d - b*c
    Dx = m*d - b*n
    Dy = a*n - m*c
    
    if D != 0:
        x = Dx / D
        y = Dy / D
        return f"Hệ có nghiệm duy nhất: x = {x}, y = {y}"
    else:
        if Dx == 0 and Dy == 0:
            return "Hệ có vô số nghiệm"
        else:
            return "Hệ vô nghiệm"

# Nhập liệu cho số thực
try:
    a, b, c, d, m, n = map(float, input('Nhap so: ').split())
    print(res := he_phuong_trinh_chi_tiet(a, b, m, c, d, n)) # 'res' ở đây là biến lưu kết quả
except ValueError:
    print("Vui lòng nhập số thực!")
```

#### Bài 10:
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
# Cach 1:
a, b = map(float, input('Nhập giá trị a, b: ').split())
r = float(input('Nhập giá trị bán kính r:'))
x, y = map(float, input('Nhập giá trị x, y: ').split())
phuong_trinh_duong_tron = (x - a)**2 + (y - b)**2 - r**2
if phuong_trinh_duong_tron < 0:
    print('Điểm nằm trong đường tròn')
elif phuong_trinh_duong_tron > 0:
    print('Điểm nằm ngoài đường tròn')
else:
    print('Điểm nằm trên đường tròn')
```

```python
# Cach 2:
def check_diem(x0, y0, r, x, y):
    dist_sq = (x - x0)**2 + (y - y0)**2
    if dist_sq == r**2:
        return "Thuoc duong tron"
    if dist_sq < r**2:
        return 'Nam trong'
    else:
        return'Nam ngoai'
try:
    x0, y0, r, x, y = map(int,input().split())
    print(f'Ket qua la: {check_diem(x0, y0, r, x, y)}')
except ValueError:
    print('Nhap dung cac diem')
```
